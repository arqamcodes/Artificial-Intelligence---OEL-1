```python
roads = {
 'S1': ['S2', 'S3'],
 'S2': ['S1', 'S4'],
 'S3': ['S1', 'S5'],
 'S4': ['S2'],
 'S5': ['S3']
}

def find_reachable_shelters(roads, start):
  queue = [start]
  visited = set()
  reachable = []

  while queue:
    shelter = queue.pop(0)
    if shelter not in visited:
      visited.add(shelter)
      reachable.append(shelter)
      neighbours = roads[shelter]
      for neighbour in neighbours:
        queue.append(neighbour)

  return reachable

find_reachable_shelters(roads, 'S1')
```
